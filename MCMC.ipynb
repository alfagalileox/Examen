# Cadenas de Markov

El m√©todo de cadenas de markov consiste en hacer caminatas aleatorias para encontrar los mejores valores que ajustan a nuestro modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data.txt")
xobs = data[:,0]
yobs = data[:,1]
```

```python
def model( xobs, m, b ):
    return m*xobs + b
    
def chi2( yobs, ymodel ):
    N = yobs.shape[0]
    p = np.sum( (yobs - ymodel)**2 )*1.0/float(N)
    return p
```
